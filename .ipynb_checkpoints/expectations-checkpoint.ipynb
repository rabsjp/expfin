### The process

\begin{align}
x_{_{t+1}} = \rho x_{_{t}} +\sigma_\epsilon\epsilon_{_{t}}
\end{align}

where $\epsilon$ is the standard normal, and $\sigma_\epsilon=20$

### Forecast
- $F_{_{t-k}}x_{_{t+1}}$ is the forecast of $x_{_{t+1}}$, $k$ periods ahead of date $t$

- $E^{FI}_{_{t-k}}x_{_{t+1}}$ is the full information rational expectation.

- The forecast is the sum of a extrapolative (overraction), $F^{e}$ and a sticky (underreaction) terms, $F^{s}$. 

- The overreaction $F^{e}$ is

\begin{align}
F^{^e}_{_{t-k}} x_{_{t+1}}= E_{_{t-k}} x_{_{t+1}} + \gamma(x_{_{t-k}}-E_{_{t-k-1}} x_{_{t-k}})
\end{align}

Extrapolative individuals react too much to unexpected innovativation ($\gamma>0$). If, however, $x$ has a deterministic tredn and does not deviate from it, subjective expectations will be rational. 

- The sticky expectations $F^{s}$ is

\begin{align}
F^s_{_{t-k}} x_{_{t+1}}= (1-\lambda) E_{_{t-k}} x_{_{t+1}} + \lambda F^s_{_{t-k-1}} x_{_{t+1}}
\end{align}

when $\lambda=1$ agents have sticky beliefs. They remain stuck with forecasts that were ratinal in the past. 


- The empirial formulation is 
\begin{align}
F_{_{t-k}} x_{_{t+1}}- E_{_{t-k}} x_{_{t+1}} = \lambda (F_{_{t-k-1}}x_{_{t+1}}-E_{_{t-k}}x_{_{t+1}})+ \gamma(x_{_{t-k}}-E_{_{t-k-1}} x_{_{t-k}})
\end{align}


- The rational expectations solutions 

\begin{align}
E^{FI}_{_{t-k}}x_{_{t}} \equiv E_{_{t-k}}x_{_{t}} = \rho^k x_{t-k}
\end{align}



### Trade-offs
Every choice involves an opportunity cost. An opportunity cost is the value of opportunities lost. By showing up to class, you are forgoing some other activity. By attending college, you are forgoing income of a full-time job in hopes that you will make enough after college to make up for this.

### Thinking on the margin 
We want to consider choices in terms of marginal costs and benefits, which help highlight the trade offs we face. 

### Trade
Better outcomes can be achieved for all when we trade. Specialization and economies of scale can help increase production.
