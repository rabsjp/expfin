- the classical paper on asset markets is due to Smith, Suchanek & Williams (Econometrica, 1988) aka SSW

- finite horizon (15 or 30 periods)

- two goods: *cash (c)*  and an intrinsically worthless *asset (s)* that pays cash **dividend (d)** each period

- d takes one of four equally probable values; e.g. {0,4,8,20}

- each Ss (9 or 12) endowed with both c and s. 

- (s) trade implement through 4-min CDA each period

- thus, $E[\sum d]$ is decreasing at a constant rate over time 

![Williams (2008)](bubble.png)Source: Williams (2008)

This paper has about 1460 citations. So the bubbles come in different flavors. for example, 

1. short-selling moderates prices, but sufficiently unconstrained leads to negative bubbles (Haruvy & Noussair, JF 2006)

2. large number of trades (see Figure 1 from William, 2008)

3. dividend certainty (Porte & Smith, J. of Business 1995)

4. Professionals. (King, Smith, Williams & Van Boening, Nonlinear dynamics 1993)

5. Experience helps reduce bubbles (King et al. 1993; Dufwenberg, Lindqvist & Moore, AER 2005), but with new parameters they bubble (Hussam, Porter & Smith, AER 2008)



The BIG question of any asset market experiment is whether the market price comes to ***agreggrate*** all of the available information.

According to theory, we should not observe any trade on SSW. Why? 
    - Everyone has the same information (common knowledge)
    - there is no real motive to trade!
    
or, if they do trade, it should occur at the fundamental value (which is observable to the experimenter) 


Bubbles appear to come from three sources 

1. Injections of liquidity (cash)

     - Caginalp et a. (2001, J Psych and Financial Markets) find that each unit of additional cash per share adds to the maximum price reached by the asset 

2. Confusion about the nature and trajectory of fundamental values 
     - (see Kirchler et al. 2012) 

3. Beliefs that other are confused about fundamental values.
